# 2. Compare Two Images

### Preparation

```python
# Import modules
import cv2
```

```python
# Load imagesa
image1 = cv2.imread('./data/cat4.png')
image1 = cv2.imread('./data/cat4.png')
image2 = cv2.resize(image, (500,500))
image2 = cv2.resize(image, (500,500))

# Change to gray scale
image1_gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
image2_gray = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
```

```python
# Rotate Image2
image2_rotated = cv2.rotate(image2, cv2.ROTATE_90_CLOCKWISE)
image2_gray_rotated = cv2.rotate(image2_gray, cv2.ROTATE_90_CLOCKWISE)
```

### Apply SIFT

```python
# Create SIFT objects
sift = cv2.SIFT_create()
print(sift)
```

```python
# Set Keypoints and Descriptors
keypoints1, descriptors1 = sift.detectAndCompute(image1_gray, None)
keypoints2, descriptors2 = sift.detectAndCompute(image2_gray_rotated, None)
print(keypoints1, keypoints2)
print(descriptors1, descriptors2)
```

### Match Keypoints

```python
# Calculate matching points
matcher = cv2.BFMatcher()
matches = matcher.match(descriptors1, descriptors2)
print(matches)
```

```python
# Sort matched points
matches = sorted(matches, key=lambda x: x.distance)
print(matches)
```

```python
# Get the Top 10 matching points
for match in matches[:10]:
    print('Distance: ', match.distance)
    print('Keypoint 1: (x=%d, y=%d)' % (int(keypoints1[match.queryIdx].pt[0]), int(keypoints1[match.queryIdx].pt[1])))
    print('Keypoint 2: (x=%d, y=%d)' % (int(keypoints2[match.trainIdx].pt[0]), int(keypoints2[match.trainIdx].pt[1])))
    print()
```

### Visualization

```python
# Visualization
image_matched = cv2.drawMatches(image1, 
                                keypoints1,
                                image2_rotated,
                                keypoints2,
                                matches[:10],
                                None,
                                flags = cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
cv2.imshow('Matched Image', image_matched)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
