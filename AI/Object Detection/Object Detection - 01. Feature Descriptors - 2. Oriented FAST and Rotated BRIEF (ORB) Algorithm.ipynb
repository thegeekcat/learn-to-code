### Preparation

```python
# Import modules
import cv2
import numpy as np
```

### Load Imaes

```python
### Create ORB objects
orb = cv2.ORB_create()
```

```python
# Load object image
image_object = cv2.imread('./data/noisy miner_22.png')
image_object = cv2.resize(image_object, (500, 500))

# Conver to gray scale
image_object_gray = cv2.cvtColor(image_object, cv2.COLOR_BGR2GRAY)

# Get Keypoints and Discroptors
keypoints_object, descriptors_object = orb.detectAndCompute(image_object_gray, None)
print(keypoints_object, descriptors_object)
```

```python
# Load target images
image_target = cv2.imread('./data/noisy miner_22.png')
image_target = cv2.resize(image_target, (500, 500))

# Conver to gray scale
image_target_gray = cv2.cvtColor(image_target, cv2.COLOR_BGR2GRAY)

# Get Keypoints and Discroptors
keypoints_target, descriptors_target = orb.detectAndCompute(image_target_gray, None)
print(keypoints_target, descriptors_target)
```

### Get Results

```python
matcher = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = matcher.match(descriptors_object, descriptors_target)
print(matches)
```

```python
# Sort matched resuts
matches = sorted(matches, key=lambda x: x.distance)
print(matches)
```

```python
# Get the Top 10 matching result
if len(matches) > 10:
    object_found = True
    
    # Get locations
    points_object = [keypoints_object[m.queryIdx].pt for m in matches]
    points_target = [keypoints_target[m.trainIdx].pt for m in matches]
    
    # Calculate axixes of bounding rectangles
    object_x, object_y, object_w, object_h = cv2.boundingRect(np.float32(points_object))
    target_x, target_y, target_w, target_h = cv2.boundingRect(np.float32(points_target))
    
    # Visualization
    cv2.rectangle(image_target, 
                  (object_x, object_y), 
                  (object_x + object_w, object_y + object_h),
                  (0, 255, 0),
                  2)
    cv2.putText(image_target, 
                'Object',
                (object_x, object_y - 10),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.9,
                (0, 255, 0),
                2)
```

```python
# Visualization
cv2.imshow('Object Recognition', image_target)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
