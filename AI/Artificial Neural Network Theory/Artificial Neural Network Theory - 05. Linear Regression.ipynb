# 1. Simple Linear Regression

```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
```

### Generate Data

```python
# Generate data
x = np.array([1, 3, 6, 9, 12])
y = np.array([2, 4, 7, 10, 13])
print('x: ', x)
print('y: ', y)
```

```python
# Visualization
plt.scatter(x, y)
plt.show()
```

### Calculate Slope and Intercept with the Least Squares Method

```python
# Calculate Slope
beta_1 = np.sum((x - np.mean(x)) * (y - np.mean(y))) / np.sum((x - np.mean(x)) ** 2)

# Calculate Intercept
beta_0 = np.mean(y) - beta_1 * np.mean(x)

print('Slope: ', beta_1)
print('Intercept: ', beta_0)

'''
Simple Linear Regression: y = B1x + B0
                            -> y = 1.0x + 1.0
'''
```

### Visualization

```python
# Display a regression line
plt.scatter(x, y)
plt.plot(x,
         beta_0 + beta_1 * x,
         color = 'red')
plt.show()
```
