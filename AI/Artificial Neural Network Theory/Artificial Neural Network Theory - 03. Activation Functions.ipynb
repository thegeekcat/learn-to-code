# 1. Sigmoid Function (=Logistic Function) 

- Sigmoid Function
  - Map input values to a range between 0 and 1 -> to provide a smooth and bounded output
  - A non-linear function
  - S(x) = 1 / (1 + e^(-x]))
  - Limitations
    - Vanishing gradient: Sigmoid function has a derivative ranging between 0 and 0.25 -> lead to slow convergence or inability of model to learn
    - Not zero-centered

```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define a function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Set values
x = np.linspace(-10, 10, 100) # Generate 100 values between -10 and 10
y = sigmoid(x)

print(x)
print(y)
print(sigmoid(100))

# Visualization
plt.plot(x, y)
plt.title('Sigmoid Function')
plt.show
```

# 2. Hyperbolic Tangent Function

- Hyperbolic Tangent Function
  - Map input values to a range between -1 and 1 -> Zero-Centered
  - A Non-Linear activation function
  - tanh(x) = (e^x - e^(-x)) / (e^x + e^(-x))

- Difference between Sigmoid and Hyperbolic Tangent
  - Sigmoid Function: Value of y is positive values
  - Hypoerbolic Tangent Function: Value of y is between 1 and -1 
  

```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define a function
def tanh(x):
    return np.tanh(x)

# Set values
x = np.linspace(-5, 5, 100) # Generate 100 values between -5 and 5
y = tanh(x)

print(x)
print(y)

# Visualization
plt.plot(x, y)
plt.title('Hyperbolic Tangent Function')
plt.show()
```

# 3. ReLU (Recitified Linear Unit) Function

- ReLU (Recitified Linear Unit) Function
  - Input values
    - Positive input values: Return the input itself
    - Negative input values: Return '0'
  - Mitigate the vanishing gradient problem 
  - Limitations
    - Dying ReLU Problem: Negative bias or large negative gradients -> get stuck in inactive state
    - Non-zero-centered output

```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define a ReLu Function
def relu(x):
    return np.maximum(0, x)

# Set values
x = np.linspace(-5, 5, 100)
y = relu(x)

print(x)
print(y)

# Visualization
plt.plot(x, y)
plt.title('ReLu Function')
plt.show()
```

# 4. Leakly ReLU Function

- Leakly ReLU Function
  - A variation of the ReLU function -> Address the 'dying ReLU' problem
  - Introduce a small slope for negative input values instead of '0'

```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define a Leaky ReLU Function
def leaky_relu (x, alpha = 0.01):
    return np.where(x >= 0,
                    x, 
                    x * alpha)

# Set values
x = np.linspace(-5, 5, 100)
y = leaky_relu(x)

print(x)
print(y)

# Visualization
plt.plot(x, y)
plt.title('Leaky ReLu Function')
plt.show()
```

# 5. Softmax Function

- Softmax Function
  - Use for multi-class classfication problems
  - Convert a vector of input values into probabilities for multiple classes
  - Normalize input values between 0 and 1 -> Total sum is '1'
  - softmax(z) = (e^z_i) / (e^z_j)

```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define a Softmax Function
def softmax(z):
    return np.exp(z) / np.sum(np.exp(z))

# Set values
z = np.linspace(-10, 10, 100)
y = softmax(z)

print(z)
print(y)

# Visualization
plt.plot(z, y)
plt.title('Softmax Function')
plt.show()
```
