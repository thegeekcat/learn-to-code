```python
# Import modules
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service

import time
import os
import urllib, requests
```

# 2. Crwaling

```python
########################################################################################
# Get a query on Chrome driver
query = 'kookaburra'
service = Service('./chromedriver')
driver = webdriver.Chrome(service = service)
driver.implicitly_wait(3)


########################################################################################
# Add query to search
driver.get('https://images.google.com/')
#driver.get('https://images.google.com/search?q=')

keyword = driver.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/textarea')
keyword.send_keys(query)


########################################################################################
# Search area: /html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/textarea
# Search button: /html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/button
# 'Show More Results' button: /html/body/div[2]/c-wiz/div[3]/div[1]/div/div/div/div/div[1]/div[2]/div[2]/input
#
# reference: driver.find_element_by_class_name('Tg7LZd').send_keys(Keys.ENTER)
########################################################################################


########################################################################################
# Search images with 'query'
driver.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/button').click()



########################################################################################
# Scrolling down
print(f'Scrolling down to search {query}!')
elem = driver.find_element_by_tag_name('body')
for i in range(60):  # Number of scrolling
    elem.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.1)
try:
    # Click 'Show more results' button by class name
    #driver.find.element_by_xpath('/html/body/div[2]/c-wiz/div[3]/div[1]/div/div/div/div/div[1]/div[2]/div[2]/input').click()
    driver.find_element_by_class_name('mye4qd').send_keys(Keys.ENTER)
    for i in range(60):
        elem.send_keys(Keys.PAGE_DOWN)
        time.sleep(0.1)
except:
    pass


########################################################################################
# Get number of images
links = []
images = driver.find_elements_by_css_selector('img.rg_i')
for image in images:
    if image.get_attribute('src') != None:
        links.append(image.get_attribute('src'))
    elif image.get_attribute('data-src') != None:
        links.append(image.get_attribute('data-src'))
    elif image.get_attribute('data-iurl') != None:
        links.append(image.get_attribute('data-iurl'))

print('Number of found images: ', len(links))


########################################################################################
# Download images
count = 0
for i in links:
    start = time.time()
    url = i
    os.makedirs(f'./data/{query}/', exist_ok=True)
    
    while True:
        try:
            urllib.request.urlretrieve(url, f'./data/{query}/{query}_{str(count)}.png')
            print(f'{str(count+1)} / {str(len(links))} / {query} / Downloaded Time: {str(time.time() - start)[:5]} secs')
            break
        except urllib.error.HTTPError as e:
            print(f'HTTP Error happened: {e}... Trying again!')
            time.sleep(5)
        except Exception as e:
            print(f'Error happened: {e}... Trying again!')
            time.sleep(5)
        
        # If time is over 60 secs, then considered as 'failed'
        if time.time() - start > 60:
            print('Failed to download {query} images :(')
            break
    count = count + 1

print('Download completed!')

driver.close()
```

```python
# Reference from 허재호님


from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By

from re import I
import time
import os 
import urllib, requests

def google_crawling(search_words, counts) :
    print("Download starts..")
    service = Service('./chromedriver')
    driver = webdriver.Chrome(service=service)
    driver.implicitly_wait(3)
    driver.get("https://www.google.co.kr/imghp?h1=ko")
    os.makedirs('./imgs/', exist_ok=True)
    current_path = os.getcwd()
    current_path += '\\imgs\\'
    for search_word in search_words:
        elem = driver.find_element(By.NAME, value="q")
        elem.clear()
        elem.send_keys(search_word)
        elem.send_keys(Keys.RETURN)
        SCROLL_PAUSE_TIME = 1
        last_height = driver.execute_script("return document.body.scrollHeight") 

        while True:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")  
            time.sleep(SCROLL_PAUSE_TIME)  
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:  
                try:
                    driver.find_elements(By.CSS_SELECTOR, value=".mye4qd").click()  
                except:
                    break
            last_height = new_height
            
        images = driver.find_elements(By.CSS_SELECTOR, value=".rg_i.Q4LuWd")
        folder_name = search_word
        if not os.path.isdir(current_path +folder_name):  
            os.makedirs(current_path +folder_name, exist_ok=True)
            
        count = 1
        for image in images:
            if count > counts:
                break
            try:
                image.click()
                time.sleep(3)
                imgUrl = driver.find_element(By.XPATH, value="/html/body/div[2]/c-wiz/div[3]/div[2]/div[3]/div[2]/div/div[2]/div[2]/div[2]/c-wiz/div/div/div/div[3]/div[1]/a/img[1]").get_attribute("src")
                urllib.request.urlretrieve(
                    imgUrl,
                    current_path  + folder_name + "/" + search_word + "_" + str(count) + ".jpg")
                count += 1
            except:
                pass
    
        driver.back()
    print("Download completed")
    driver.close()

input_count = int(input("다운받을 목록 개수를 입력하세요: "))
input_list = []
for i in range(input_count):
    input_value = input("목록 {} 입력하세요: ".format(i + 1))
    input_list.append(input_value)
print("목록 :", input_list)
counts = int(input("다운받을 이미지 개수를 입력하세요: "))
google_crawling(input_list, counts)
```
