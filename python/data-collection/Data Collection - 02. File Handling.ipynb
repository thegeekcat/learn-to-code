# 1. Get a List of Files: `os.listdir()`

- `os.listdir()`: No order

```python
# Import modules
import os
```

```python
# Set a path
img_dir = './data/noisy miner/'

# Get a list of files in the directory: `os.listdir()
file_list = os.listdir(img_dir)
print(file_list)
```

# 2. Get a List of Files Sorted: `sorted()`

- `sorted()`
  - Only sort alphabets

```python
# Import modules
import os
```

```python
# Set a path
img_dir = './data/noisy miner/'

# Get a list of files in the directory: `sorted(os.listdir())`
file_list = sorted(os.listdir(img_dir))
print(file_list)
```

# 3. Get a List of Files Sorted: `natsort.nasorted()`

- natsort.natsort()
  - Able to order alphabets + numbers

```python
# Install modules
!pip install natsort
```

```python
# Import modules
from natsort import natsort
```

```python
# Set a path
img_dir = './data/noisy miner/'

# Get a list of files in the directory: `nasorted(os.listdir())`
file_list = natsort.natsorted(os.listdir(img_dir))
print(file_list)
```

# 4. Get a List of Files with Paths: `glob.glob()`

- `glob.glob()`
  - Get a list of files with their folder paths

```python
# Import modules
import glob
import os
```

```python
file_list = glob.glob(os.path.join('./data/noisy miner/', '*.png'))
print(file_list)
```

```python
# Sort it
file_list = natsort.natsorted(glob.glob(os.path.join('./data/noisy miner/', '*.png')))
print(file_list)
```

# 5. : `os.walk()`

```python
# Import modules
import os
```

```python
# 


def get_img_paths(root_path): # `root_path`: Set a root path
    file_paths = []
    
    # Get info of path, directories, files
    for (path, dir, files) in os.walk(root_path):
        #print('Path: ', path)
        #print('Dir: ', dir)
        #print('Files: ', files)
        
        # Get a list of files with paths
        for file in files:
            
            
            # Get file name extensions
            ext = os.path.splitext(file)[-1].lower()   # '-1': to make sure to get the last extension
            #print('File name extensions: ', ext)
            
            # Set a list of graphical files
            formats_list = ['.bmp', '.jpg', '.jpeg', '.png', '.tif', '.gif', '.dng', '.tiff']
            
            # Geta list of image files only
            if ext in formats_list:
                file_path = os.path.join(path, file)
                #print('File Path: ', file_path)
                file_paths.append(file_path)

        return file_paths
            
    
file_path = get_img_paths("./data/noisy miner")
#print(file_path)

# Get a list of files
for i in file_path:
    print(i)
```
