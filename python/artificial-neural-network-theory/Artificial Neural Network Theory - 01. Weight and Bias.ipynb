# 1. Weight

```python
# Import modules
import torch
import matplotlib.pyplot as plt
```

```python
# Initialize randomly 2-dimentional weight matrix
w = torch.rand(3, 3)
print('Original Tensor: ', w)

# Change Rank2 Tensor to Rank1 Tensor for visualization
w_flattened = w.flatten()
print('Flattened Tensor', w_flattened)
```

```python
# Visualization
plt.hist(w_flattened, bins = 20)
plt.xlabel('Weight Value')
plt.ylabel('Frequency')
plt.show()
```

# 2. Bias

```python
# Initialize Bias value
bias = torch.rand(1)
print(bias)

# Visualization
plt.hist(bias.detach().numpy(), # `detach().numpy()`: Convert the bias tensor to a NumPy array
         bins = 20)  # Higher value: See roughly, 
plt.xlabel('Bias Value')
plt.ylabel('Frequency')
plt.show()
```
