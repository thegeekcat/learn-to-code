### Preparation

```python
# Import modules
import cv2
```

### Load Images

```python
# Load images
image1 = cv2.imread('./data/cat9-2.jpg')
#image2 = cv2.imread('./data/cat4.png')
image2 = cv2.imread('./data/cat9-3.jpg')

# Resize images
image1 = cv2.resize(image1, (500, 500))
image2 = cv2.resize(image2, (500, 500))

# Rotate images
image2 = cv2.rotate(image2, cv2.ROTATE_90_CLOCKWISE)
```

### Create ORB Objects

```python
orb = cv2.ORB_create()
print(orb)
```

### Get Keypoints and Descriptors

```python
keypoints1, descriptors1 = orb.detectAndCompute(image1, None)
keypoints2, descriptors2 = orb.detectAndCompute(image2, None)
print(keypoints1, descriptors1)
print(keypoints2, descriptors2)
```

### Find Matches

```python
# Get Matching points
matcher = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = matcher.match(descriptors1, descriptors2)
print(matches)
```

```python
# Sort matched results
matches = sorted(matches, key=lambda x: x.distance)
print(matches)
```

### Visualization

```python
image_matched = cv2.drawMatches(image1,
                                keypoints1,
                                image2,
                                keypoints2,
                                matches[:20],
                                None,
                                flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
cv2.imshow('BRIEF Matching', image_matched)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
