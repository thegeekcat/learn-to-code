# 1. Image Blending

```python
# Import modules
import cv2
import matplotlib.pyplot as plt
```

```python
# Load images
sample_img = cv2.imread('./data/ex_image.png')
sample_logo = cv2.imread('./data/ex_image_logo.png')

sample_img = cv2.cvtColor(sample_img, cv2.COLOR_BGR2RGB)
sample_logo = cv2.cvtColor(sample_logo, cv2.COLOR_BGR2RGB)

# Visualization
plt.imshow(sample_img)
plt.title('Sample Image')
plt.show()

plt.imshow(sample_logo)
plt.title('Sample Logo')
plt.show()
```

```python
# Resize images
img1 = cv2.resize(sample_img, (800, 600))
img2 = cv2.resize(sample_logo, (800, 600))
print(img1.shape, img2.shape)

# Visualization
plt.imshow(img1)
plt.title('Resized Sample Image')
plt.show()

plt.imshow(img2)
plt.title('Resized Sample Logo')
plt.show()
```

```python
# Blend images: `cv2.addWeighted()`
blended = cv2.addWeighted(img1, 
                          0.5,  # Alpha
                          img2, 
                          0.7,  # Beta
                          0)    # Gamma
# Result: (img1 * Alpha) + (img2 * Beta) + Gamma

# Visualization
plt.imshow(blended)
plt.title('Blended Image')
plt.show()
```

# 2. Bitwise Operation

### Prepration

```python
# Import modules
import numpy as np
```

```python
# Generate a rectangle
img_rectangle = np.ones((400, 400), dtype='uint8')
cv2.rectangle(img_rectangle,
              (50, 50),        # Position of Top-Left corner
              (300, 300),      # Position of Bottom-Right conner
              (255, 255, 255), # Color
              -1)              # Thickness

# Visualization
plt.imshow(img_rectangle, 'gray')
plt.show()
```

```python
# Generate a circle
img_circle = np.ones((400, 400), dtype='uint8')
cv2.circle(img_circle,
           (300, 300), # Center
           50,         # Radius
           (255, 255, 255),
           -1)

# Visualization
plt.imshow(img_circle, 'gray')
plt.show()
```

### AND(A, B)

```python
bitwiseAnd = cv2.bitwise_and(img_rectangle, img_circle)
plt.imshow(bitwiseAnd, 'gray')
plt.title('AND(A, B)')
plt.show()
```

### OR(A, B)

```python
bitwiseOr = cv2.bitwise_or(img_rectangle, img_circle)
plt.imshow(bitwiseOr, 'gray')
plt.title('OR(A, B)')
plt.show()
```

### XOR(A, B)

```python
bitwiseXor = cv2.bitwise_xor(img_rectangle, img_circle)
plt.imshow(bitwiseXor, 'gray')
plt.title('XOR(A, B)')
plt.show()
```

### NOT(A)

```python
bitwiseNot_rec = cv2.bitwise_not(img_rectangle)
plt.imshow(bitwiseNot_rec, 'gray')
plt.title('NOT(A)')
plt.show()
```

### NOT (B)

```python
bitwiseNot_cir = cv2.bitwise_not(img_circle)
plt.imshow(bitwiseNot_cir, 'gray')
plt.title('NOT(B)')
plt.show()
```

### Masking Area

```python
# Generate a canvas
img_rectangle = np.ones((500, 1000), dtype='uint8')
mask = np.zeros(img_rectangle.shape, dtype='uint8')

# Add additional rectangles
cv2.rectangle(mask, (60, 50), (280, 280), (255, 255, 255), -1)
cv2.rectangle(mask, (420, 50), (550, 230), (255, 255, 255), -1)
cv2.rectangle(mask, (750, 50), (920, 280), (255, 255, 255), -1)

# Visualization
plt.imshow(mask, 'gray')
plt.show()
```
