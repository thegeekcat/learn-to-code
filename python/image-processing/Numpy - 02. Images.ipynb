```python
# Import modules
import numpy as np

import cv2
from matplotlib import pyplot as plt
```

```python
# Load an image file
img_org = cv2.imread('./data/cat.png')

img_org
```

```python
# Convert BGR to RGB
img_converted = cv2.cvtColor(img_org, cv2.COLOR_BGR2RGB)

img_converted
```

```python
# Save images as numpy
np.save('./data/image_org.npy', img_org)
np.save('./data/image_converted.npy', img_converted)
```

# 2. Comparison of Efficiency

### Preparation

```python
# Install modules
!pip install Pillow
```

```python
# Import modules
import numpy as np

from PIL import Image
```

### Load Images

```python
# Load image with grey scaling
img = Image.open('./data/cat.png').convert('L') 

img
```

### Convert Image to Numpy Array

```python
img_array = np.array(img)

img_array
```

### Save and Load as Numpy format

```python
# Save as a numpy file
np.save('./data/cat_grey', img_array)
```

```python
# Load the file
img_array = np.load('./data/cat_grey.npy')

img_array
```

### Convert Array to Image

```python
# Convert an array to an image
img = Image.fromarray(np.uint8(img_array))

# Save as an image
img.save('./data/cat_grey.png')
```

### Compare Sizes

```python
# Import modules
import os
```

```python
# Compare sizes
np_save_size = os.path.getsize('./data/cat_grey.npy')
pil_save_size = os.path.getsize('./data/cat_grey.png')

print(f'A file size of "nupy.save()": {np_save_size} bytes')
print(f'A file size of "PIL.Image.fromarray()": {pil_save_size} bytes')
```
