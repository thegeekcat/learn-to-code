# 1. Add Noise

### Preparation

```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
```

### Load Files

```python
# Load a sound file
file_path = './data/0620-sound_data/blues/blues.00000.wav'
data, sr = sf.read(file_path)

print('Data: ', data)
print('Sampling Rate: ', sr)
```

### Augmentation: Generate Noise

```python
# Generate noise
noise_amp = 0.5 * np.random.uniform() * np.amax(data)   # 'np.amax()': Return the max of array
noise = noise_amp * np.random.normal(size = data.shape)

print('noise_amp: ', noise_amp)
print('noise: ', noise)
```

```python
# Add noise to the existing data
data_noise = data.astype('float32') + noise.astype('float32')
```

### Export as a Sound File

```python
# Save as a file
save_path = './outcomes/0620-noise-added-sample-file.wav'
sf.write(save_path, data_noise, sr)
```

### Visualization

```python
# Visualization
plt.figure(figsize = (20, 5))

# Original Sound File
plt.subplot(1, 2, 1)
plt.plot(data)
plt.title('Original Sound File')

# Noise Added Sound File
plt.subplot(1, 2, 2)
plt.plot(data_noise)
plt.title('Noise Added Sound File')

plt.show()
```

# 2. Stretching

### Preparation

```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
import librosa
import soundfile as sf
```

### Load Sound Files

```python
# Load a sound file
data, sr = librosa.load('./blues.00000.wav')
print('Data: ', data)
print('Sampling Rate: ', sr)
```

### Augmentation: Stretching

```python
data_stretch = librosa.effects.time_stretch(data, rate=0.8) # 'rate=0.8': Stretch 80% of the original file
print(data_stretch)
```

### Save as a File

```python
# Save as a file
save_path = './0620-stretched.wav'
sf.write(save_path, data_stretch, sr)
```

### Visualization



```python
# Visualization
plt.subplot(1, 2, 1)

# Original Sound Image
plt.subplot(1, 2, 1)
librosa.display.waveshow(data, sr=sr, alpha=0.8)
plt.title('Original Image Sound')

# Augmentation Added Sound Image
plt.subplot(1, 2, 2)
librosa.display.waveshow(data_stretch, sr=sr, alpha=0.8)
plt.title('Stretched Image Sound')

plt.tight_layout()
plt.show()
```
