## 4.1. Preparation

```python
# Import modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate Time Series Data
np.random.seed(85)

dates = pd.date_range(start='2023-01-01', periods=1000, freq='D')
values = np.random.randn(1000)

dates, values
```

```python
# Create a dataframe
data = pd.DataFrame({
    'Date': dates,
    'Values': values
})
# 
#pd.set_option('display.max_row', 90000)
#pd.set_option('display.max_columns', 100)
data
```

## 4.2. Line Plot

```python
# Visualization
plt.plot(data['Date'], data['Values'])

plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Time Series - Line Plot')
plt.legend()
plt.xticks(rotation=45)

plt.show()
```

- Pivot Table
  : Summarize data by grouping them by specific columns
  
e.g.
1. Original Data  
Date      value   
23-01-01    A  
23-01-01    B  
23-01-02    A  
23-01-02    C  
23-01-03    B  
23-01-03    C   
   
2. Pivot Table   
          A   B   C    
23-01-01  A   B  NAN    
23-01-02  A  NAN  C   
23-01-03 NAN  B   C   

```python
# Change datatype as Pivot table
pivot_table = data_heatmap.pivot(index='Date', columns='Values', values='Values')

pivot_table
```

```python
# Visualization
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, cmap='YlOrRd', cbar='True')
plt.xlabel('Values')
plt.ylabel('Date')

plt.show()
```
