# 1. Normal Distribution

```python
# Import modules
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Generate data
np.random.seed(42)
height = np.random.normal(loc=170, scale=10, size=100)
weight = np.random.normal(loc=70, scale=5, size=100)

print('Height: \n', height)
print('\nWeight: \n', weight)
```

```python
# Change data to DataFrame
data = pd.DataFrame({'Height': height, 'Weight': weight})

data
```

```python
# Export as a csv file
os.makedirs("./data", exist_ok=True)
data.to_csv('./data/human_data.csv', index=False)
```

```python
# Check the saved file
data = pd.read_csv('./data/human_data.csv')

data
```

```python
# Get data from DataFrame
height_data = data['Height']
weight_data = data['Weight']
```

```python
# Visualization: Histogram
plt.hist(height, bins=10, alpha=0.5, label='Height')  # alpha: Level of transparancy
plt.hist(weight, bins=10, alpha=0.5, label='Weight')
plt.legend(loc='upper right')
plt.show()
```

```python
# Normalization for Height
height_mean, height_std = norm.fit(height_data)

height_x = np.linspace(height_data.min(), height_data.max(), 100)
height_y = norm.pdf(height_x, height_mean, height_std)

print(height_x, '\n\n', height_y)
```

```python
# Normalization for Weight
weight_mean, weight_std = norm.fit(weight_data)

weight_x = np.linspace(weight_data.min(), weight_data.max(), 100)
weight_y = norm.pdf(weight_x, weight_mean, weight_std)

print(weight_x, '\n\n', weight_y)
```

```python
# Visualization for Normalization
plt.hist(height, bins=10, density=True, alpha=0.5, label='Height')
plt.plot(height_x, height_y, label='Height Distribution')
plt.hist(weight, bins=10, density=True, alpha=0.5, label='Weight')
plt.plot(weight_x, weight_y, label='Weight Distribution')
plt.legend(loc='upper right')
plt.show()
```

# 2. Binomial Distribution

```python
# Import modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
# Bionomial Distribution Modelling
n = 10
p = 0.5
```

```python
# Bionomial Distribution simulation
num_simulations = 1000

successes = np.random.binomial(n, p, num_simulations)

successes
```

```python
# Visualization
plt.hist(successes, bins=11, range=(0, 10), alpha=0.5)

plt.xlabel('Success')
plt.ylabel('Frequency')

plt.title('Binomial Distribution Simulation')

plt.show()
```

# 4. Exponential Distribution

```python
# Define a Exponential Distribution generation function for modeling Waiting time
def generate_waiting_time(lam, size):
    data = np.random.exponential(scale=1/lam, size=size)
    print(data)
    return data

# Test
generate_waiting_time(0.5, 1000)
```

```python
# Define a Exponential Distribution generation function for distribution and visualization
def measure_waiting_time(lam, num_samples): 
    waiting_times = generate_waiting_time(lam, num_samples)
    # print(waiting_times)

    # Histogram
    plt.hist(waiting_times, bins=30, density=True, alpha=07.)
    plt.xlabel('Waiting Time')
    plt.ylabel('Probability Density')
    plt.title('Customer Service Center Waiting Time Distribution')
    plt.show()
    
# Test
measure_waiting_time(0.5, 100)
```

# 6. Beta Distribution

```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

# 8. T Distribution

```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set parameters

df = 10  # Degree of freedom
num_simulations = 1000
t_values = np.random.standard_t(df, num_simulations)

t_values
```

```python
# Visualization: Histogram
plt.hist(t_values, bins=20, alpha=0.5)
plt.xlabel('T-Value')
plt.ylabel('Frequency')
plt.title('T Distribution Simulation')
plt.show()
```

# 9. F Distribution

```python
# Import modules
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# F Distribution modeling
dfn = 5   # Degree of Freedom 1
dfd = 10  # Degree of Freedom 2
```

```python
# F Distribution simulations
num_simulations = 100

f_values = np.random.f(dfn, dfd, num_simulations)

f_values
```

```python
# Visualization: Histogram
plt.hist(f_values, bins=20, alpha=0.5)
plt.xlabel('F-Value')
plt.ylabel('Frequency')
plt.title('F Distribution Simulation')
plt.show()
```
