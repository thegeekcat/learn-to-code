```python
# Import modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# 2. Basics

```python
# Load a file
data = pd.read_csv("/AISchoolData/DataAnalysis/BlackFriday.csv")
print(data)
```

```python
# Sum a total amount of 'Purchase' by 'UserID'
user_total_spent = data.groupby('User_ID')["Purchase"].sum()
print(user_total_spent)
```

```python
# Create a histogram
plt.hist(user_total_spent, bins=50)
plt.title('Histogram of Total Purchase Amount by User')
plt.xlabel('Total Purchase Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a Density Plot
plt.figure()
user_total_spent.plot.kde()
plt.title('Density Plt of Total Purchase Amount by User')
plt.xlabel('Total Purchase Amount')
plt.show()
```

# 2. Normalization

```python
# Generate normally distributed data
data = np.random.normal(loc=0, scale=1, size=1000)
print(data)
```

```python
# Visualization
sns.histplot(data)
plt.show()
```

```python
# Calculate Mean and SD
mean = np.mean(data)
std = np.std(data)

print("Mean: ", mean)
print("Standard Deviation: ", std)
```

# 3. Central Tendency

```python
# Generate data
data = np.random.normal(loc=0, scale=1, size=10000)

print(data)
```

```python
# Calculate Central Tendency
mean = np.mean(data)
median = np.median(data)
mode = np.round(np.mean(data))
```

```python
# Visualization
fig, ax = plt.subplots()

ax.hist(data, bins=50)

ax.axvline(mean, color='red', linestyle='dashed', linewidth=2)
ax.axvline(median, color='green', linestyle='dashed', linewidth=2)
ax.axvline(mode, color='purple', linestyle='dashed', linewidth=2)

ax.legend(['Mean: {:.2f}'.format(mean), 'Median: {:.2f}'.format(median), 'Mode: {}'.format(mode)])

ax.set_xlabel('Data Values')
ax.set_ylabel('Frequency')
ax.set_title('Data Distribution')

plt.show()
```

# 4. Degree of Scattering

```python
# Create a relationship between two values
np.random.seed(0)

x = np.random.normal(loc=0, scale=1, size=100)
y = x + np.random.normal(loc=0, scale=0.5, size=100)
# The reason of adding 'x' to y -> to generate 'y' data similar to 'x'
#  -> This is a wrong way, but it is used for excercise here
```

```python
# Visualization

fig, ax = plt.subplots()

ax.scatter(x,y)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Scatter Plot')

plt.show()
```

# 5. Outliers

```python
# Generate data

# np.concatenate(): Combine arrays
data = np.concatenate([np.random.normal(0, 1, 900),
                      np.random.normal(8, 2, 100),
                      np.random.normal(-4, 2, 100)])

print(data)
```

```python
# Visualization: Box Plot
fig, ax = plt.subplots()

ax.boxplot(data)

ax.set_xlabel('Data')
ax.set_ylabel('Values')
ax.set_title('Box Plot')

plt.show()
```

```python
# Visualization: Histogram -> Check central tendency
fig, ax = plt.subplots()

ax.hist(data, bins=50)

ax.set_xlabel('Data')
ax.set_ylabel('Frequency')
ax.set_title('Histogram')
plt.show()
```

# 6. Correlation Coefficient

```python
# Generate data
np.random.seed(0)

x = np.random.normal(loc=0, scale=1, size=100)
y = 2*x + np.random.normal(loc=0, scale=1, size=100)
# Reason of add '2x' to 'y': Just to get a better graph to understand the concept of correlation

print(y)
```

```python
# Visualization: 

fig, ax = plt.subplots()

ax.scatter(x,y)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Scatter Plot')

plt.show()
```

```python
# Calculate Correlation Coefficient
corr = np.corrcoef(x, y)[0, 1]
print("Correlation Coefficient: ", corr)
```
