# 4. Linear and Non-Linear Relationship

```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate a linear relationship
x_linear = np.linspace(0, 10, 100)
y_linear = 2 * x_linear + 1

print('X Linear: ', x_linear)
print('\nY Linear: ', y_linear)
```

```python
# Generate a Non-Linear Relationship
x_nonlinear = np.linspace(-10, 10, 100)
y_nonlinear = x_nonlinear ** 2

print('X Non-Linear: ', x_nonlinear)
print('\nY Non-Linear: ', y_nonlinear)
```

```python
# 선형 관계 시각화
plt.scatter(x_linear, y_linear)
plt.title('Linear Relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# 비선형 관계 시각화
plt.scatter(x_nonlinear, y_nonlinear)
plt.title('Nonlinear Relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

# 6. Time Series Correlation Analysis

```python
# Import modules
import pandas as pd
import random
import os
import matplotlib.pyplot as plt
```

```python
# Generate data: 'dates'
dates = pd.date_range(start='2021-01-04',
                       periods = 100,  # Numbers of data
                       freq = 'D')

dates
```

```python
# Generate data: `Samsung`
samsung_prices = [random.randint(80000, 90000) for _ in range(100)]
samsung_data = {'Date': dates,
               '005930.KS': samsung_prices}
samsung_df = pd.DataFrame(samsung_data)

samsung_df
```

```python
# Clean `Samsung` data: Remove Indexes
samsung_df.set_index('Date', inplace=True)  # Replace 'Date' as an Index column

samsung_df
```

```python
# Generate data: `LG`
lg_prices = [random.randint(140000, 160000) for _ in range(100)]
lg_data = {'Date': dates, '066570.KS': lg_prices}
lg_df = pd.DataFrame(lg_data)

lg_df
```

```python
# Clean 'LG' data: Remove Indexes
lg_df.set_index('Date', inplace=True)

lg_df
```

```python
# Combine two dataframes: `pd.concat()`
df = pd.concat([samsung_df, lg_df], axis=1)

df
```

```python
# Rename columns
df = df.loc[:, ['005930.KS', '066570.KS']]

df.columns = ['Samsung', 'LG']

df
```

```python
# Create a folder
os.makedirs("./data", exist_ok=True)

# Save as a new file
df.to_csv("./data/stock_price.csv")
```

```python
# Load the file
df = pd.read_csv('./data/stock_price.csv')

df
```

```python
# Change 'Date' column to 'datetime' type
df['Date'] = pd.to_datetime(df['Date'])

# Replace 'Date' as an Index
df.set_index('Date', inplace=True)

df
```

```python
# Calculate returns: `df.pct_change()`
returns = df.pct_change()  

returns
```

```python
# Calculate the Correlation Coefficient between returns
corr_matrix = returns.corr()

corr_matrix
```

```python
# Visualization: Heapmap
sns.heatmap(corr_matrix, 
           annot = True,
           cmap = 'coolwarm')

plt.show()
```
