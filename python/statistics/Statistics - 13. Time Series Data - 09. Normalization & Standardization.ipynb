# 09. Normalization & Standardization

## 09.1. Min-Max Normalization

### Preparation

```python
# Import modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate data
np.random.seed(85)

date = pd.date_range(start='2023-01-01', periods=100, freq='D')
values = np.random.randint(0, 100, size=100)

date, values
```

```python
# Creae a dataframe
data = pd.DataFrame({
    'Date': date,
    'Values': values
})

data
```

### Min-Max Normalization

```python
# Get Min&Max values
min_value = data['Values'].min()
max_value = data['Values'].max()

min_value, max_value
```

```python
# Calcualte normalization
data['Value_Normalized'] = (data['Values'] - min_value) / (max_value - min_value)

data
```

```python
# Import modules
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
# Generate data
np.random.seed(85)

date = pd.date_range(start='2023-01-01', periods=100, freq='D')
values = [10, 12, 11, 15, 9, 13, 8, 14, 10, 16] * 10

date, values
```

```python
# Create a dataframe
data = pd.DataFrame({
    'Date': date,
    'Values': values
})

data
```

### Standardization

```python
# Get X
X = data['Values'].values.reshape(-1, 1)

#X
```

```python
# Standardization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled
```

```python
# Add standardized X to a new column
data['Value_scaled'] = X_scaled

data
```
