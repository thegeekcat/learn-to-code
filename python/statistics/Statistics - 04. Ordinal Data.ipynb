```python
# Import modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Basics

```python
# Load data
data = pd.read_csv('./data/college_data.csv')

data.head()
```

```python
# Devide Top 10 percent students into 4 groups: `pd.qcut'

# `pd.qcut`: Devide data equally
data['adminssion_level'] = pd.qcut(data['top10perc'], 
                                   q=4,   # Numbers of groups
                                   labels=['very_low', 'low', 'high', 'very_high'])  # Labels of groups
```

```python
# Visualization
plt.figure(figsize=(10, 6))
plt.bar(data['admission_level'].value_counts().index, data['admission_level'].value_counts().values)
plt.title('Number of Colleges by Admission Level')
plt.xlabel('Admission Level')
plt.ylabel('Number of Colleges')
plt.show()
```
