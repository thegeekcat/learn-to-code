### Preparation

```python
# Import modules
import numpy as np
from sklearn.decomposition import FactorAnalysis
```

```python
# Generate data
np.random.seed(85)
n_samples = 100
n_features = 5

X = np.random.rand(n_samples, n_features)
X
```

### Factor Analysis

```python
# Factor Analysis
n_components = 2   # Numbers of components

# Generate 
factor_analysis = FactorAnalysis(n_components = n_components,  
                                random_state = 85) # Random Seed

X_factors = factor_analysis.fit_transform(X)
```

```python
# Result
print("The size of original dataset: ", X.shape)
print("\nResult of Factor Analysis(Number of Factors: {})".format(n_components))
print('\n', X_factors)
```
